### Time To model

```python
# DataTraining = pd.read_csv('../../data/processed/flights_final_processed.csv')
df
```

```python
x = df.drop('ARRIVAL_DELAY', axis=1)
y = df['ARRIVAL_DELAY']
```

```python
x.columns
```

```python
# Scaling all the continous data 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_scale = ["MONTH", "DAY", "DAY_OF_WEEK", "SCHEDULED_TIME", "DEPARTURE_DELAY","DISTANCE"]
x[columns_to_scale] = scaler.fit_transform(x[columns_to_scale])
```

```python

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.2, random_state=42
)
```

```python
INPUT_DIM = 21
UNITS = [512, 256, 128, 64, 32]  
BATCH_SIZE = 256
EPOCHS = 20
LR = 1e-4
```

```python
model = keras.Sequential()
model.add(layers.Input(shape=(INPUT_DIM,)))

for u in UNITS:
    model.add(layers.Dense(u, activation='elu'))

model.add(layers.Dense(1, activation='linear'))
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss='mean_squared_error',
    metrics=['mae', 'mse']
)
```

```python
model.fit(x_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, validation_data=(x_val, y_val))
```

```python
# test the model accuracy
model.evaluate(x_test, y_test)
```

```python
# print accuracy
y_pred = model.predict(x_test)
```

```python
# evaluate the model by comparing y_test and y_pred and give me precentae accuracy

from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```

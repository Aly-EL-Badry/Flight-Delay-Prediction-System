### Time To model

```python
# DataTraining = pd.read_csv('../../data/processed/flights_final_processed.csv')
df
```

```python
x = df.drop('ARRIVAL_DELAY', axis=1)
y = df['ARRIVAL_DELAY']
```

```python
x.columns
```

```python
# Scaling all the continous data 
from sklearn.preprocessing import StandardScaler

x["MONTH_sin"] = np.sin(2*np.pi * x["MONTH"]/12)
x["MONTH_cos"] = np.cos(2*np.pi * x["MONTH"]/12)

x["DAY_sin"] = np.sin(2*np.pi * x["DAY"]/31)
x["DAY_cos"] = np.cos(2*np.pi * x["DAY"]/31)

x["DOW_sin"] = np.sin(2*np.pi * x["DAY_OF_WEEK"]/7)
x["DOW_cos"] = np.cos(2*np.pi * x["DAY_OF_WEEK"]/7)

x = x.drop(["MONTH", "DAY", "DAY_OF_WEEK"], axis=1)

scaler = StandardScaler()
columns_to_scale = ["SCHEDULED_TIME", "DEPARTURE_DELAY","DISTANCE"]
x[columns_to_scale] = scaler.fit_transform(x[columns_to_scale])
```

```python

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.2, random_state=42
)
```

```python
INPUT_DIM = x_train.shape[1]
BATCH_SIZE = 1024
EPOCHS = 30
LR = 3e-4
```

```python
model = keras.Sequential([
    layers.Input(shape=(INPUT_DIM,)),

    layers.Dense(256),
    layers.BatchNormalization(),
    layers.Activation('swish'),

    layers.Dense(128),
    layers.BatchNormalization(),
    layers.Activation('swish'),
    layers.Dropout(0.05),

    layers.Dense(64),
    layers.BatchNormalization(),
    layers.Activation('swish'),
    layers.Dropout(0.05),

    layers.Dense(1)
])
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss='mae',
    metrics=['mse', 'mae']
)
```

```python
early = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=15, restore_best_weights=True
)
```

```python
model.fit(x_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, validation_data=(x_val, y_val),callbacks=[early])
```

```python
# test the model accuracy
model.evaluate(x_test, y_test)
```

```python
# print accuracy
y_pred = model.predict(x_test)
```

```python
# evaluate the model by comparing y_test and y_pred and give me precentae accuracy

from sklearn.metrics import mean_squared_error, mean_absolute_error , r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_percent = r2 * 100
print(f"R^2 Score: {r2}")
print(f"Accuracy Percent: {accuracy_percent}%")
print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```
